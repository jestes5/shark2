Precipitation data is collected in 15-minute increments from over 3000 sites in the United States of America. We want to look at data from Salt Lake city and from Provo city to see how small differences in region impact overall rainfall. Let's take a look!

Let's look at each year's rainfall using the following code:

```python
print("year   rainfall")
for i in range(len(years)):
  print(years[i],":",annual_precipitation[i])
```

Now let's find the year with the maximum rainfall and the year with the minimum rainfall:
 
*(note: the minimum value excludes any zero values assuming that those are incorrect.)*

```python
max_precip = 0
min_precip = 2000
print("     year   rainfall")
for i in range(len(years)):
  if annual_precipitation[i] > max_precip:
    max_precip = annual_precipitation[i]
    max_year = years[i]
  if annual_precipitation[i] < min_precip and annual_precipitation[i] != 0:
    min_precip = annual_precipitation[i]
    min_year = years[i]
print("max ", max_year, ":", max_precip)
print("min ", min_year, ":", min_precip)
```
